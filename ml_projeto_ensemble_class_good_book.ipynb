## vamos começar o tratamento de dados

transformar a coluna review/helpfulness	em um float


```python

# Separando a coluna pelo delimitador /
novas_colunas = df['review/helpfulness'].str.split('/', expand=True)

# Renomeando as novas colunas para melhor clareza
novas_colunas.columns = ['numerator-helpfulness', 'denominator-helpfulness']

# Adicionando as novas colunas ao DataFrame original
df = pd.concat([df, novas_colunas], axis=1)
```

```python
df
```

```python
# convertendo as novas colunas para float
df['numerator-helpfulness'] = df['numerator-helpfulness'].astype(float)
df['denominator-helpfulness'] = df['denominator-helpfulness'].astype(float)
```

```python
# obtendo os valores do quanto aquela avaliação foi útil
df['fraction-helpfulness'] = df['numerator-helpfulness'] / df['denominator-helpfulness']
```

```python
df['fraction-helpfulness'].describe()
```

```python
df.drop(['numerator-helpfulness', 'denominator-helpfulness', 'review/helpfulness'], axis = 1, inplace = True)
```

```python
df.info()
```

## Agora, temos apenas colunas textuais, vamos transformar em colunas de vetores

```python
# converter todas as letras para minusculas

colunas_textuais = ['title', 'review/summary', 'review/text','description'
,'authors', 'categories']

for coluna in colunas_textuais:
  df[coluna] = df[coluna].str.lower()
```

```python
df
```

```python
# remover pontuação
import regex as re

def remover_pontuação(texto):
  """
  remove pontuação usando regex
  \p{P} é uma propriedade unicode que corresponde a qualquer tipo de pontuação
  """
  if isinstance(texto, str):
    return re.sub(r'\p{P}', '', texto, flags=re.UNICODE)
  return texto

df[colunas_textuais] = df[colunas_textuais].apply(lambda x: x.apply(remover_pontuação))
```

```python
df
```

```python
# vamos agora remover as stopwords

colunas_textuais = ['title', 'review/summary', 'review/text','description'
,'authors', 'categories']

import nltk
from nltk.corpus import stopwords
import string

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def remover_stop_words(texto):
  if isinstance(texto, str):
    palavras = texto.split()

    palavras_filtradas = [
        palavra for palavra in palavras
        if palavra not in stop_words
    ]

    return " ".join(palavras_filtradas)
  return "" # Return an empty string for non-string inputs

df[colunas_textuais] = df[colunas_textuais].apply(lambda x: x.apply(remover_stop_words))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler

# Um TF-IDF separado para cada coluna
text_transformers = [
    (f"text_{col}", TfidfVectorizer(), col)
    for col in colunas_textuais
]

# transformador_numerico
numeric_transformer = StandardScaler()
```

```python
from sklearn.compose import ColumnTransformer

colunas_numericas = ['fraction-helpfulness', 'price']

# pre-processamento de transformação das colunas
preprocessor = ColumnTransformer(
    transformers = (
        text_transformers +
        [("num", numeric_transformer, colunas_numericas)]
    ),
    remainder="drop"
)
```

testando os ensemble methods

```python
# voting classifier - softing vote (pelas probabilidades)
from sklearn.ensemble import VotingClassifier
from sklearn.pipeline import Pipeline

voting_clf = Pipeline([
    ("preprocessor", preprocessor),
    ("voting", VotingClassifier(
        estimators=[
            ('knn', knn_base),
            ('lr', lr_base),
            ('dt', dt_base)
        ],
        voting='soft'
    ))
])
```

```python
voting_clf.fit(X_train, y_train)
```

```python
# stacking -> Treina um modelo final (meta-model) sobre as previsões dos modelos base.
from sklearn.ensemble import StackingClassifier

stacking_clf = Pipeline([
    ("preprocessor", preprocessor),
    ("stack", StackingClassifier(
        estimators=[
            ('knn', knn_base),
            ('lr', lr_base),
            ('dt', dt_base)
        ],
        final_estimator=LogisticRegression(),
        cv=5,                         # obrigatório para gerar preds out-of-fold
        stack_method="predict_proba", # garante uso das probas como features do meta
    ))
])
```

```python
stacking_clf.fit(X_train, y_train)
```

avaliando os modelos de montagem criados

```python
from sklearn.metrics import classification_report, accuracy_score, roc_auc_score

def avaliar(modelo, X_test, y_test, nome):
    y_pred = modelo.predict(X_test)
    print(f"\n===== {nome} =====")
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("ROC AUC Score:", roc_auc_score(y_test,  modelo.predict_proba(X_test)[:,1]))
    print(classification_report(y_test, y_pred))
```

```python
avaliar(voting_clf, X_test, y_test, "VotingClassifier")
avaliar(stacking_clf, X_test, y_test, "StackingClassifier")
```

Com isso, podemos ver que nem sempre os ensemble methods se saem melhor que os modelos mais simples.

Se sairam melhor que os modelos KNN e DT, no entanto não superaram os resultados obtidos pela regressão logística
